# Matplotlib figure 图形对象
在 Matplotlib 中，面向对象编程的核心思想是创建图形对象（figure object）。通过图形对象来调用其它的方法和属性，有助更好地处理多个画布。

在这个过程中，pyplot 负责生成图形对象，并通过该对象来添加一个或多个 axes 对象（即绘图区域）。

Matplotlib 提供了 `matplotlib.figure` 图形类模块，它包含了创建图形对象的方法。

通过调用 pyplot 模块中 figure() 函数来实例化 figure 对象：

## 创建画布对象

```python
import matplotlib.pyplot as plt
# 创建图形对象
fig = plt.figure()
```

其构造函数的参数如下：
- `figsize`：指定画布的大小，(宽度,高度)，单位为英寸。
- `dpi`：指定绘图对象的分辨率，即每英寸多少个像素，默认值为80。
- `facecolor`：背景颜色。
- `dgecolor`：边框颜色。
- `frameon`：是否显示边框。

## 添加轴

使用 `add_axes` 方法，将轴添加到对象中。

方法的参数是序列，序列中的四个数字分队对应图形的左侧，底部，宽度，和高度，且每个数字必须介于 0 到 1 之间。

方法返回添加进图形中的轴对象

```python
# 给 fig 对象添加轴
ax = fig.add_axes([0, 0, 1, 1])
```

## 设置轴的标签

```python
ax.set_title('my wave')
ax.set_xlabel('angle')
ax.set_ylabel('sin')
```

## 绘制轴
调用 `axes` 对象的 `plot` 方法

```python
import numpy as np
import math
x = np.arange(0, math.pi * 2, 0.02)
y = np.sin(x)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y)
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')
plt.show()
```
