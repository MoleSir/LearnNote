# Matplotlib 双轴图
Matplotlib 提供的 `twinx()` 和 `twiny()` 函数，可以实现绘制双轴的功能，并且还可以使用不同的单位来绘制曲线，比如一个轴绘制对函数，另外一个轴绘制指数函数。
>如果在一个 `axes` 对象中直接多次 `plot`，这些曲线都只能使用同一套坐标，要多个轴，需要多个 `axes`
>

下面示例绘制了一个具有两个 y 轴的图形，一个显示指数函数 exp(x)，另一个显示对数函数 log(x)

```python
import matplotlib.pyplot as plt
import numpy as np
import math

fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])

x = np.arange(0, 2 * math.pi, 0.1)
y1 = np.exp(x)
y2 = np.log(x)

ax1.plot(x, y1,)
ax1.set_ylabel('exp')

# 添加双轴
ax2 = ax1.twinx()
ax2.plot(x, y2, 'ro-')
ax2.set_ylabel('log')

fig.legend(labels = ('exp', 'log'), loc = 'upper left')

plt.show()
```
